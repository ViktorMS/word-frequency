```
import pandas as pd
df = pd.read_csv('spam.csv')
df.head()
```

```
df.describe()
```

```
msg = df.Message
msg_split = msg.str.split(expand=True)
msg_split.head()
```

```
msg_stack = msg_split.stack()
msg_stack.head()
```

```
n_most_frequent_words = 100

msg_word_frequency = msg_stack.value_counts()
msg_word_frequency.head(n_most_frequent_words)
```
